Here, we outline the qiime2 and python commands used to process the Chagas disease datasets using Gneiss 0.4.0

```bash

for f in aq_annot org_annot aq_full org_full; do
    qiime tools import --input-path $f.biom \
                       --output-path $f.biom.qza \
                       --type FeatureTable[Composition]

    qiime gneiss correlation-clustering \
                        --i-table $f.biom.qza \
                        --o-clustering $f.proportional_tree.nwk.qza \
                        --verbose

    qiime gneiss ilr-transform \
                        --i-table $f.biom.qza \
                        --i-tree $f.proportional_tree.nwk.qza \
                        --o-balances $f.balances.qza

    qiime gneiss ols-regression \
                        --p-formula "BURDEN + POSITION + DOSE + OUTCOME" \
                        --i-table $f.balances.qza \
                        --i-tree $f.proportional_tree.nwk.qza \
                        --m-metadata-file "$f"_metadata.txt \
                        --o-visualization $f.parasites_summary
done
```

## Part 1: look at the OUTCOME and position for aq

Based on the coefficient heatmap in `aq_annot.parasites_summary.qzv`, y0 could be interesting.   First let's load the result datasets.

````python
import pandas as pd
import qiime2
import numpy as np
mapping = pd.read_table('aq_annot_metadata.txt', index_col=0)
balances = qiime2.Artifact.load('aq_annot.balances.qza').view(pd.DataFrame)
````

Here, we will plot `y0` with respect to the OUTCOME and the DOSE.

````python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
%matplotlib inline

plt.rc('text', usetex=False)
def transform_outcome(x):
    if x['DOSE'] == 1000:
        x['OUTCOME'] = '50% FATAL'
    return x
mapping.index = mapping.index.astype(np.str)
data = pd.merge(mapping, balances, left_index=True, right_index=True)

data = data.apply(transform_outcome, axis=1)
fig = sns.factorplot(x='DOSE', y='y0', hue='OUTCOME', data=data, kind='box', size=4, aspect=2,
                     palette={'FATAL': '#FF0000','NON_FATAL': '#008000', 'UNINFECTED': '#0000FF', 
                              '50% FATAL': '#F27304'})
fig.savefig('dose.pdf')
fig.savefig('dose.png')
````

Let's also take a look at `y2`, since that also appears to be interesting.  We'll use a bag-of-words approach to try to make sense of the balances.  This will involve analyzing the subwords within each of the annotations, and observing which chemical compound names appear most frequently.

````python
import re
def bag_of_words(X):
    """
    Parameters
    ----------
    X : list of str
        List of identifications
    
    Returns
    -------
    bag: pd.Series
        Counts of words        
    """
    pattern = re.compile('([a-z A-z]+)')
    bag = []
    for x in X:
        p = pattern.findall(x)
        bag += list(filter(lambda x: len(x)>3, p))
    bag = pd.Series(bag)
    bag = bag.value_counts()
    bag.name = 'molecules'
    return bag
````

Now we'll load up the partitions, and extract the numerator and denominator compounds for `y2`.

````python
from skbio import TreeNode
from gneiss.util import NUMERATOR, DENOMINATOR

tree = qiime2.Artifact.load('aq_annot.proportional_tree.nwk.qza').view(TreeNode)

num = list(tree.find('y2').children[NUMERATOR].subset())
denom = list(tree.find('y2').children[DENOMINATOR].subset())
````

We will only to plot a summary of the top 8 most commonly observed compounds, so we'll trim the counts as follows.

````python
def trim_counts(x, n):
    x_ = x.copy()
    other = x_.iloc[n:].sum()
    x_ = x_.head(n)
    x_['other'] = other
    x_ = x_.reset_index()
    return x_

top_n = 8
num_bag = trim_counts(
    bag_of_words(num), top_n)
denom_bag = trim_counts(
    bag_of_words(denom), top_n)
````

Let's view the boxplot, along with barplot summaries of the numerator and the denominator.  Here we will only focus on the high dose samples.

````python
sns.set(font_scale=2) 
sns.set_style("white")

high_dose_data = data.loc[data.DOSE==1000000]

plt.rc('text', usetex=True)
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 3))
gs = gridspec.GridSpec(100, 100)
ax1 = plt.subplot(gs[:, :20])
ax2 = plt.subplot(gs[:, 40:60])
ax3 = plt.subplot(gs[:, 80:])

sns.boxplot(hue='OUTCOME', y='POSITION', x='y2', data=high_dose_data, ax=ax1, 
            order=['A', 'B', 'C', 'D'], 
            palette={'FATAL': '#FF0000','NON_FATAL': '#008000'})
ax1.legend_.remove()
ax2.set_title(r"$y2_{numerator}$")
sns.barplot(y='index', x='molecules', data=num_bag, ax=ax2, color='#FF0000')
ax3.set_title(r"$y2_{denominator}$")
sns.barplot(y='index', x='molecules', data=denom_bag, ax=ax3, color='#008000')


ax1.set_title(r"y2=$\displaystyle\log(\frac{y2_{numerator}}{y2_{denominator}})$",
              fontsize=15)

ax1.set_xlabel("magnitude")
ax1.set_ylabel("")
ax2.set_xlabel("number of IDs")
ax2.set_ylabel("")
ax3.set_xlabel("number of IDs")
ax3.set_ylabel("")

ax1.locator_params(axis='x', nbins=4)
ax2.locator_params(axis='x', nbins=4)
ax3.locator_params(axis='x', nbins=4)
f.savefig('1000000_fatal-nonfatal.png')
f.savefig('1000000_fatal-nonfatal.pdf')
````

Now's try to analyze both the high dose and low dose samples.

````python
from gneiss.util import match

sns.set(font_scale=2) 
sns.set_style("white")
# filter out uninfected outcomes
mapping = mapping.loc[mapping.OUTCOME != 'UNINFECTED']
data, mapping = match(data, mapping)

high_dose_data = data.loc[data.DOSE==100000]
low_dose_data = data.loc[data.DOSE==100]
dose_data = pd.concat((high_dose_data, low_dose_data))

plt.rc('text', usetex=True)
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 3))
gs = gridspec.GridSpec(100, 100)
ax1 = plt.subplot(gs[:, :20])
ax2 = plt.subplot(gs[:, 40:60])
ax3 = plt.subplot(gs[:, 80:])

sns.boxplot(hue='OUTCOME', y='POSITION', x='y2', data=dose_data, ax=ax1, 
            order=['A', 'B', 'C', 'D'], 
            palette={'FATAL': '#FF0000','NON_FATAL': '#008000'})
ax1.legend_.remove()
ax2.set_title(r"$y2_{numerator}$")
sns.barplot(y='index', x='molecules', data=num_bag, ax=ax2, color='#FF0000')
ax3.set_title(r"$y2_{denominator}$")
sns.barplot(y='index', x='molecules', data=denom_bag, ax=ax3, color='#008000')


ax1.set_title(r"y2=$\displaystyle\log(\frac{y2_{numerator}}{y2_{denominator}})$",
              fontsize=15)

ax1.set_xlabel("magnitude")
ax1.set_ylabel("")
ax2.set_xlabel("number of IDs")
ax2.set_ylabel("")
ax3.set_xlabel("number of IDs")
ax3.set_ylabel("")

ax1.locator_params(axis='x', nbins=4)
ax2.locator_params(axis='x', nbins=4)
ax3.locator_params(axis='x', nbins=4)
f.savefig('100_vs_100000_fatal-nonfatal.png')
f.savefig('100_vs_100000_fatal-nonfatal.pdf')
````
